```python
##Preprocessing
##One hotEncoding with get dumies#
```

```python
X_dumm = pd.get_dummies(X_cat, drop_first=False) # You can put X, not separating variables is needed.
X_dumm
```

```python
##Concat
X_num = X.select_dtypes(include = np.number)
X_concat= pd.concat([X_num, X_dumm], axis=1)
X_concat.sample(10)
```

```python
##Train Test Split(Define Train test for X and Y)##
### test_size =  30% for testing and 70% for training
##Test size is the ratio of original data which we can use as a test set
# random_state = it'll improve the model to divide always the model in the same way
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
## Apply Linear Regression
```

```python
model = LinearRegression()## Model is linear Regression
model.fit(X_train,y_train)## Fit the training data in to Model defined
```

```python
##Model Validation
predictions  = model.predict(X_test)
```

```python
predictions ## It should give an array of predictions
```

```python
y_test ## array of data from our target
```

```python
##Model Validation
##Description:
R2.
MSE.
RMSE.
MAE.
```

```python
r2 = r2_score(y_test, predictions)
RMSE = mean_squared_error(y_test, predictions, squared=False)
MSE = mean_squared_error(y_test, predictions)
```

```python
adj_r2 = (1 - (1 - r2) * ((X_concat.shape[0] - 1) / 
          (X_concat.shape[0] - X_concat.shape[1] - 1)))
print("r2 = ", r2)
print("RMSE = ", RMSE)
print("MSE = ", MSE)
print("Adj r2 = ", adj_r2)
```
